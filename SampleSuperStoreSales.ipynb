```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Sample Superstore Dataset.csv',encoding='latin1')
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.dropna()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.strip()
```

```python
df.head()
```

```python
df['profitmargin']=(df['profit']/df['sales'])*100
```

```python
df['revenue']=df['sales']*(1-df['discount'])
```

```python
df.columns=df.columns.str.replace(' ','')
```

```python
df.columns
```

```python
df.head()
```

```python
import matplotlib.pyplot as py
import seaborn as sns
```

```python
df.groupby('orderdate')['sales'].sum().plot(figsize=(20,5))
py.title('sales over time')
py.show()
```

```python
sns.barplot(x='region',y='sales',data=df)
py.title('Region sales')
py.show()
```

```python
df.to_csv('Sample Superstore Dataset.csv',index='false')
```
